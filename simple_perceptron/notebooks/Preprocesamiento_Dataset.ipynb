```python
import pandas as pd
import os

# Subimos un nivel desde src → llegamos a la raíz del proyecto
base_dir = os.path.join('..', 'data')

raw_path = os.path.join(base_dir, 'raw', 'iris.data')
processed_path = os.path.join(base_dir, 'processed', 'iris_binario.csv')

# Leemos archivo
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df = pd.read_csv(raw_path, names=column_names)

# Filtramos clases
df_binario = df[df['class'] != 'Iris-virginica'].copy()

# Guardamos
df_binario.to_csv(processed_path, index=False)

print(f"Éxito: {processed_path}")
```
